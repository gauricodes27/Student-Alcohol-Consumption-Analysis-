```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('/content/student-mat.csv')
df2=pd.read_csv('/content/student-por.csv')
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
# Age vs Alcohol Consumption
sns.barplot(data=df, x='age', y='Walc', ci=None)
plt.title("Weekend Alcohol Consumption by Age")
plt.show()
```

```python
# Gender vs Alcohol Consumption
sns.boxplot(data=df, x='sex', y='Walc')
plt.title("Weekend Alcohol Consumption by Gender")
plt.show()
```

```python
# Histogram of Walc by address
sns.histplot(data=df, x='Walc', hue='address', multiple='dodge', bins=5, shrink=0.8)
plt.title("Histogram of Weekend Alcohol Consumption: Urban vs Rural")
plt.xlabel("Weekend Alcohol Consumption (Walc)")
plt.ylabel("Count")
plt.legend(title='Address', labels=['Rural (R)', 'Urban (U)'])
plt.show()
```

```python
# Family Relationship Quality
sns.barplot(data=df, x='famrel', y='Walc', ci=None)
plt.title("Family Relationship Quality vs Alcohol")
plt.show()
```

```python
# Q: What age is most prone to get into drinking habits?
# Switched to a violin plot to show distribution and density
sns.violinplot(data=df, x='age', y='Walc', palette='magma')
plt.title("Age vs Weekend Alcohol Consumption")
plt.show()
```

```python
# Q: Which gender drinks more?
# Switched to countplot for distribution clarity across alcohol levels
sns.countplot(data=df, x='Walc', hue='sex', palette='Set2')
plt.title("Gender-wise Alcohol Consumption Distribution")
plt.xlabel("Weekend Alcohol Consumption")
plt.ylabel("Count")
plt.show()
```

```python
# Q: Subject analysis – Which subject’s students drink more?
df['subject'] = ['Math'] * len(df1) + ['Portuguese'] * len(df2)
sns.violinplot(data=df, x='subject', y='Walc', palette='coolwarm')
plt.title("Subject vs Weekend Alcohol Consumption")
plt.show()
```

```python
# Q: Do students with less educated parents consume more alcohol?
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(data=df, x='Medu', y='Walc', ci=None, ax=axs[0], palette='Blues')
axs[0].set_title("Mother's Education vs Alcohol")
axs[0].set_xlabel("Medu (0 = none, 4 = higher ed)")
sns.barplot(data=df, x='Fedu', y='Walc', ci=None, ax=axs[1], palette='Greens')
axs[1].set_title("Father's Education vs Alcohol")
axs[1].set_xlabel("Fedu (0 = none, 4 = higher ed)")
plt.tight_layout()
plt.show()
```

```python
# Q: How does the type of parental job influence alcohol consumption?
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=df, x='Mjob', y='Walc', ci=None, ax=axs[0], palette='flare')
axs[0].set_title("Mother's Job vs Alcohol")
axs[0].tick_params(axis='x', rotation=45)
sns.barplot(data=df, x='Fjob', y='Walc', ci=None, ax=axs[1], palette='crest')
axs[1].set_title("Father's Job vs Alcohol")
axs[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q10: What kind of relationship does the child have with family members?
sns.barplot(data=df, x='famrel', y='Walc', ci=None, palette='Spectral')
plt.title("Family Relationship Quality vs Alcohol Consumption")
plt.xlabel("Family Relationship Quality (1 = very bad, 5 = excellent)")
plt.show()
```

```python
# Q11: Family background’s influence on alcoholism (correlation heatmap)
corr_data = df[['Medu', 'Fedu', 'famrel', 'Walc']].copy()
corr_data['famsup_num'] = df['famsup'].map({'yes': 1, 'no': 0})  # Convert to numeric

corr = corr_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation between Family Background & Alcohol Consumption")
plt.show()
```

```python
# Q: Golden balance group (moderate goout, moderate studytime, low alcohol use, high grades)
df['Sum/60'] = df[['G1', 'G2', 'G3']].sum(axis=1)
sns.scatterplot(data=df, x='goout', y='studytime', size='Sum/60', hue='Walc', palette='coolwarm', alpha=0.7)
plt.title("Golden Balance: Studytime vs Goout vs Alcohol vs Grades")
plt.xlabel("Go Out Frequency")
plt.ylabel("Study Time")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
df_num = df.select_dtypes(include=['int64', 'float64'])  # only numeric cols
df_num['avg_alcohol'] = (df['Dalc'] + df['Walc']) / 2

plt.figure(figsize=(12, 8))
sns.heatmap(df_num.corr()[['avg_alcohol']].sort_values('avg_alcohol', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Alcohol Consumption')
plt.show()
```

```python
# Assuming 'Dalc' (daily) and 'Walc' (weekend) alcohol consumption scores are already in your df
df['avg_alcohol'] = (df['Dalc'] + df['Walc']) / 2

# Create a categorical alcohol level column
df['Alcohol_Level'] = pd.cut(
    df['avg_alcohol'],
    bins=[0, 1.5, 3.5, 5],
    labels=['Low', 'Moderate', 'High']
)
```

```python
sns.boxplot(data=df, x='Alcohol_Level', y='academic_engagement_score', palette='cool')
plt.title('How Alcohol Affects Academic Engagement')
plt.show()
```
